## Executive Summary

This reports presents a thorough examination of weekly U.S. All Grades All Formulations Retail Gasoline Prices in Dollars per Gallon. The dataset provides crucial insights into the U.S. retail gasoline market, aiding in understanding trends and guiding decision-making.

Initial data preprocessing involved testing for stationarity using the Augmented Dickey-Fuller (ADF) test, revealing non-stationarity (ADF Statistic: -2.5478, p-value: 0.1042). Differencing was then applied to the data to make it stationary which was reinforced by the positive findings in the ADF test. The absence of a significant drop in the autocorrelation function suggested no moving average component, while a notable drop in the partial autocorrelation function indicated the presence of autoregression.

### Testing for Stationarity

The first step in the analysis is to assess the stationarity of the US Gasoline Retail Prices dataset using the Augmented Dickey-Fuller (ADF) test. Stationarity is a fundamental assumption for many time series models, as it implies that the statistical properties of the data, such as mean and variance, remain constant over time.

The ADF test is a statistical method used to determine whether a time series is stationary or non-stationary. It examines the presence of unit roots in the data, which indicate non-stationarity.

**> ADF Test Results:**

The results of the ADF test indicate that the US Gasoline Retail Prices dataset is not stationary.

*   Test Statistic: -2.5478030071370674
*   p-value: 0.10427218376736413


First-order differencing was applied to the US Gasoline Retail Prices dataset by computing the difference between each observation and its preceding observation:
$$
\Large y'_t= y_t-y_{t-1}
$$

Where ' represents the differenced series at time , and   represents the original series at time .


**> Impact of Differencing:**

Differencing removes the trend component from the data, thereby stabilizing the mean and making the series approximately stationary. By stabilizing the mean, differencing facilitates the application of time series models that assume stationarity. The effects of differencing can be visualised in the below chart.

##Is it Moving Average?

An ACF plot was generated for the differenced series to visualize the correlation between observations at different lags. The ACF plot helps determine whether there is a correlation between an observation and its lagged values. If the autocorrelation coefficients at various lags are significantly different from zero, it suggests the presence of autocorrelation in the data.

image.png



> In the case of this dataset (seen above) the lags progressively decreased indicating that there was no significant auto correlation which means that the moving average model is not applicable.



## Is  it Auto-Regression
In addition to the autocorrelation function (ACF) plot, an analysis of the partial autocorrelation function (PACF) was generated to further understand the temporal dependencies within the US Gasoline Retail Prices dataset. The PACF plot helps identify the direct relationship between observations at different lags while controlling for the indirect effects of intermediate lags.

image.png



> Based on the insights from the PACF plot, we determined that an autoregressive (AR) model with order 3 is appropriate for modeling the US Gasoline Retail Prices dataset. The significant spikes at specific lags in the PACF plot indicate a direct relationship between observations at those lags, supporting the choice of an AR model with three lag terms.



Since it has been recognised that the data is third-order autoregressive process, or $AR(3)$, the present value $y_t$ is
linearly dependent on a constant $C$, the value at the previous timestep $\phi_1 y_{t 1}$, the value two timesteps prior $\phi_2 y_{t 2}$, the value three timesteps prior $\phi_3 y_{t 3}$,and the present error term $\epsilon_t$, as shown in equation:

$$
\Large y_t = C + \phi_1 y_{t 1} + \phi_2 y_{t 2} + \phi_3 y_{t 3} + \epsilon_t
$$
